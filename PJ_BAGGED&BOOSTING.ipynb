```python
X = AAPL[['Open','High','Low','Close','Adj Close']]
X1 = AAPL[['Open','High','Low','Close','Adj Close','Volume']]
x_train = X[:500]
x1_train = X1[:500]
y = AAPL[['target']]
y = y.values.ravel()
y_train = y[:500]
x_test = X[500:]
x1_test = X1[500:]
y_test = y[500:]
```

```python
# Bagged Decision Trees for Classification
import pandas
from sklearn import model_selection
from sklearn.ensemble import BaggingClassifier
from sklearn.tree import DecisionTreeClassifier
seed = 123
kfold = model_selection.KFold(n_splits=10)
cart = DecisionTreeClassifier()
num_trees = 100
#model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=seed)
#results = model_selection.cross_val_score(model,x_train,y_train,cv=kfold)
#model = BaggingClassifier(base_estimator=cart, n_estimators=num_trees, random_state=seed)
#results = model_selection.cross_val_score(model,x_train,y_train,cv=kfold)
#print(results.mean())

clfbc = BaggingClassifier(base_estimator=cart,n_estimators=10,random_state=7).fit(x_train,y_train)
for i in range(1,3599):
  y_pred = clfbc.predict(x_test[:i])

y_pred
```

```python
y_pred = pd.DataFrame(y_pred)
a = y_pred
a
```

```python
a = a.rename(columns = {0:'x'})
a
```

```python
for i in range(0,500):
  a.loc[i] = ['NaN']
  i += 1
  
a = a[:500]
a
```

```python
Y = pd.concat([a,y_pred],axis = 0).reset_index()
Y[:505]
#Y.tail()
#Y.shape
#Y
```

```python
AAPL['Bagg'] = Y[0]
#AAPL[505:550]
AAPL[:505]
```

```python
# GradientBoosting
import pandas
from sklearn import model_selection
from sklearn.ensemble import GradientBoostingClassifier

num_trees = 100
#kfold = model_selection.KFold(n_splits=20)
#model = GradientBoostingClassifier(n_estimators=num_trees, random_state=seed)
#results = model_selection.cross_val_score(model,x_train,y_train,cv=kfold)
#print(results.mean())
clfgb = GradientBoostingClassifier(n_estimators=100, learning_rate=1.0,max_depth=1, random_state=0).fit(x_train, y_train)

for i in range(1,3599):
  y_pred1 = clfgb.predict(x_test[:i])

y_pred1
```

```python
y_pred1 = pd.DataFrame(y_pred1)
b = y_pred1
b = b.rename(columns = {0:'x'})
for i in range(0,500):
  b.loc[i] = ['NaN']
  i += 1
  
b = b[:500]

Y1 = pd.concat([b,y_pred1],axis = 0).reset_index()
AAPL['Boosting'] = Y1[0]
```
